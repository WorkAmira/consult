### DIMENSIONALITY REDUCTION

<h4> VESSEL_OWNER's

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = "Other"
    return categorical_map
```

```python
owner_map = shorten_categories(df_vendor["VESSEL_OWNER"].value_counts(), 400)
df_vendor["VESSEL_OWNER"] = df_vendor["VESSEL_OWNER"].map(owner_map)
```

### OUTLIER REMOVAL

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_vendor['QUOTE_AMOUNT'].quantile(0.25)
Q3 = df_vendor['QUOTE_AMOUNT'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_vendor = df_vendor[(df_vendor['QUOTE_AMOUNT'] >= lower_bound) & (df_vendor['QUOTE_AMOUNT'] <= upper_bound)]
```

```python
def remove_outliers(df, column, group_by_col):
    def iqr_filter(group):
        Q1 = group[column].quantile(0.25)
        Q3 = group[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        return group[(group[column] >= lower_bound) & (group[column] <= upper_bound)]
    
    return df.groupby(group_by_col).apply(iqr_filter).reset_index(drop=True)

df_vendor = remove_outliers(df_vendor, 'QUOTE_AMOUNT', 'PORT_NAME')

print(f"Original data size: {df_vendor.shape[0]}")
print(f"Cleaned data size: {df_vendor.shape[0]}")
```

### MODEL TRAINING

```python
X = df1.drop(["DISCOUNT","CREATE_DATE", "QUOTE_AMOUNT", "TOTAL_COST", "VESSEL_OWNER"], axis=1)  # Features
Y = df1["TOTAL_COST"]   # Label
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
y_pred = linear_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
print("${:,.02f}".format(error))
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(X_train, y_train)
```

```python
y_pred = dec_tree_reg.predict(X_test)
```

```python
error = np.sqrt(mean_squared_error(y_test, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(X_train, y_train)
```

```python
y_pred = random_forest_reg.predict(X_test)
```

```python
error = np.sqrt(mean_squared_error(y_test, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2,4,6,8,10,12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=20)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(X_train, y_train)
```

```python
regressor = gs.best_estimator_

regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
error = np.sqrt(mean_squared_error(y_test, y_pred))
print("${:,.02f}".format(error))
```

### EXPORTING PICKLE FILE

```python
import pickle

# Export the model to a pickle file
with open('model.pkl', 'wb') as file:
    pickle.dump(regressor, file)
```

```python
# Load the model from the pickle file
with open('model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

y_loaded_pred = loaded_model.predict(X_test)
loaded_error = np.sqrt(mean_squared_error(y_test, y_loaded_pred))
print("${:,.02f}".format(loaded_error)) 
```
